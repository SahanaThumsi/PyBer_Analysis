```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# set the x Axis to a list of strings for each month
x_axis  = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
# set the y axis to a list of floats as total fare in us dollars accumulated for each month 
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis, y_axis)
```

```python
# create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot (x_axis, y_axis)
```

```python
# create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, marker="D", color="green", linewidth=2, label='Boston')
ax.legend()
ax.grid()
ax.set_title("Pyber Fare By Month")
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.set_ylim(0, 45)

```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# create a plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston' )
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid
plt.grid()
# Add the legend.
plt.legend()
```

```python
# create the plot
plt.bar(x_axis, y_axis, color="green", label='Boston')
# create the label for x axis and y axis 
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
#Add the legend
plt.legend()
```

```python
# Craete the plot
plt.barh(x_axis, y_axis, color="magenta", label='Boston')
plt.gca().invert_yaxis()
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
#Add the legend
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i * 5 for i in y_axis], color="skyblue", label= 'Boston', alpha=0.8, edgecolor="black", linewidth=2)
ax.legend()
# adding lables
ax.set_ylabel("Date")
ax.set_xlabel("Fare($)")
# Create a  title
ax.set_title("PyBer Fare By Month")
ax.invert_yaxis()
```

```python
# Assign 12 colors, one for each month.
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.subplots(figsize=(8, 8))
plt.pie(y_axis,
    explode=explode_values,
    colors=colors,
    labels=x_axis,
    autopct='%.1f%%')

plt.show()
```
